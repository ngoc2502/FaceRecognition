```python
from PIL import Image
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np
```

Take data from camera and save to folders. Each folder hold a set of images for each person

```python
video=cv2.VideoCapture(0)
facedetect=cv2.CascadeClassifier('E:/UwayInternshipLearning/FaceRecognition/haarcascade_frontalFace_default.xml')
count=0

nameID=str((input('name: ')).lower())
path='images/'+nameID

Exist=os.path.exists(path)

if Exist:
  print('Exist!')
  nameID=str(input("Enter name: "))
else:
  os.makedirs(path)

while True:
  ret,frame=video.read()
  faces=facedetect.detectMultiScale(frame)
  for x,y,w,h in faces:
    count+=1
    name='./images/'+nameID+'/'+str(count)+'.jpg'
    print('.....',name)
    cv2.imwrite(name,frame[y:y+h,x:x+w])
    cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),3)
  cv2.imshow('windowFrame ',frame)
  cv2.waitKey(1)
  if count>300:
    break
video.release()
cv2.destroyAllWindows()
```

Create path of each image and save to "Faces_path" (Faces_path[0]->ngoc, Faces_path[1]->Linh,...)

```python
from pandas import DataFrame

path='images'

labels=[]
for labl in os.listdir("images"):
    labels.append(labl)

paths=[]
for i in  labels:
    paths.append(os.path.join(path,i))

faces_path=[]
col=['image','name']

# df=DataFrame(columns=col)
# dic_img={}

for p in range (len(paths)):
    label_name=paths[p][paths[p].find('images/')+7:]
    for dirname,_,names in os.walk(paths[p]):
        image_path=[]
        for name in names:
            image_path.append(os.path.join(dirname,name))
    faces_path.append(image_path)
# t=5
# print(labels[t])
# faces_path[t]
# df''
```

```python
train_X=[]
train_Y=[]
for i in range(len(faces_path)):
    for j in faces_path[i]:
        im=Image.open(j).convert("L")
        im=im.resize((144,144))
        arr_im=np.array(im,"uint8")
        train_X.append(arr_im)
        train_Y.append(i)
     
        
```
